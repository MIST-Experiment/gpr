```python
import numpy as np
import matplotlib.pyplot as plt

# Chirp parameters
center_freq = 1e9  # 1 GHz
K = 50e6  # Chirp rate in Hz/s
duration = 1e-3  # 1 ms

# Time vector
t = np.linspace(0, duration, int(duration * 100e6))  # 100 MHz sampling rate

# Generate complex chirp signal with K parameter
chirp_signal = np.exp(2j * np.pi * (center_freq * t + 0.5 * K * t**2))

# Matched filter design (complex conjugate of the chirp signal)
matched_filter = np.conj(chirp_signal)

# Compute autocorrelation using convolution (matched filtering)
autocorr_result = np.convolve(chirp_signal, matched_filter, mode='full')

# Time axis for autocorrelation result
autocorr_time = np.linspace(-duration, duration, len(autocorr_result))

# Plot the original chirp signal
plt.figure(figsize=(12, 6))

plt.subplot(2, 1, 1)
plt.plot(t, np.real(chirp_signal), label='Real Part')
plt.plot(t, np.imag(chirp_signal), label='Imaginary Part')
plt.title('Complex Chirp Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()

# Plot the matched filter
plt.subplot(2, 1, 2)
plt.plot(t, np.real(matched_filter), label='Real Part')
plt.plot(t, np.imag(matched_filter), label='Imaginary Part')
plt.title('Matched Filter (Complex Conjugate of Chirp Signal)')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()

plt.tight_layout()
plt.show()

# Plot the autocorrelation result
plt.figure(figsize=(12, 4))
plt.plot(autocorr_time, 10*np.log10(np.abs(autocorr_result)))
plt.title('Chirp Signal Autocorrelation (Matched Filter)')
plt.xlabel('Time Lag (s)')
plt.ylabel('Autocorrelation Magnitude')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, ifft

# Chirp parameters
center_freq = 1e9  # 1 GHz
K = 50e6  # Chirp rate in Hz/s
duration = 1e-3  # 1 ms

# Time vector
t = np.linspace(0, duration, int(duration * 100e6))  # 100 MHz sampling rate

# Generate complex chirp signal with K parameter
chirp_signal = np.exp(2j * np.pi * (center_freq * t + 0.5 * K * t**2))

# Matched filter design (complex conjugate of the chirp signal)
matched_filter =(chirp_signal)

# Perform FFT on the chirp signal and the matched filter
chirp_signal_fft = fft(chirp_signal)
matched_filter_fft = fft(matched_filter)

# Multiply the frequency domain representations (convolution in time domain)
autocorr_result_fft = chirp_signal_fft * np.conj(matched_filter_fft)

# Perform IFFT to get back to the time domain
autocorr_result = ifft(autocorr_result_fft)

# Time axis for autocorrelation result
autocorr_time = np.linspace(-duration, duration, len(autocorr_result))

# Plot the original chirp signal
plt.figure(figsize=(12, 6))

plt.subplot(2, 1, 1)
plt.plot(t, np.real(chirp_signal), label='Real Part')
plt.plot(t, np.imag(chirp_signal), label='Imaginary Part')
plt.title('Complex Chirp Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()

# Plot the matched filter
plt.subplot(2, 1, 2)
plt.plot(t, np.real(matched_filter), label='Real Part')
plt.plot(t, np.imag(matched_filter), label='Imaginary Part')
plt.title('Matched Filter (Complex Conjugate of Chirp Signal)')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()

plt.tight_layout()
plt.show()

# Plot the autocorrelation result
plt.figure(figsize=(12, 4))
plt.plot(autocorr_time, np.abs(autocorr_result))
plt.title('Chirp Signal Autocorrelation (Frequency Domain Matched Filter)')
plt.xlabel('Time Lag (s)')
plt.ylabel('Autocorrelation Magnitude')
plt.grid(True)
plt.show()
```

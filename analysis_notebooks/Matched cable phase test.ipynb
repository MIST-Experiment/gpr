```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
```

```python
dat05 = np.fromfile(open("phase_0.8m_50.5MHz.dat"), dtype="float32")
dat1 = np.fromfile(open("phase_0.8m_51MHz.dat"), dtype="float32")
dat15 = np.fromfile(open("phase_0.8m_51.5MHz.dat"), dtype="float32")
dat2 = np.fromfile(open("phase_0.8m_52MHz.dat"), dtype="float32")
```

```python
phase05 = np.average(dat05)
phase1 = np.average(dat1)
phase15 = np.average(dat15)
phase2 = np.average(dat2)
```

```python
def phase_estimator(len_diff, freq, vf):
    return (2*np.pi*len_diff*freq)/(vf*299792458)
```

```python
phase05_est = phase_estimator(0.9, 50.5e6, 0.7)
phase1_est = phase_estimator(0.9, 51e6, 0.7)
phase15_est = phase_estimator(0.9, 51.5e6, 0.7)
phase2_est = phase_estimator(0.9, 52e6, 0.7)
```

```python
phase05_err = phase05_est - phase05
phase1_err = phase1_est - phase1
phase15_err = phase15_est - phase15
phase2_err = phase2_est - phase2
```

```python
f, ax = plt.subplots()

ax.scatter([0.5, 1, 1.5, 2], [phase05_err, phase1_err, phase15_err, phase2_err], label = "50MHz")

ax.set_xlabel('Signal Frequency (MHz)')
ax.set_ylabel('Phase error')

plt.title('Errors for matching cables')

plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as signal
```

```python
#dat = np.fromfile(open("beat_fmcw_uhd.dat"), dtype="float32")
#dat = np.fromfile(open("lowpass_test.dat"), dtype="complex64")
dat = np.fromfile(open("/home/mhetu/Desktop/gpr/gnu_log/out0_log"), dtype="complex64")
```

```python
plt.specgram(dat, NFFT=1024, Fs=12000000)
plt.title("PSD of reflected sigal")
plt.xlabel("Time")
plt.ylabel("Frequency")
plt.show() 
```

```python
plt.psd(dat[215000:217000], NFFT=1024, Fs=12000000)
plt.title("PSD of reflected signal")
plt.show() 
```

```python
clock = np.arange(len(dat))
plt.plot(clock[800000:1400000], dat[800000:1400000])
```
